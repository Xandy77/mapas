Na aula de hoje, vamos trabalhar com **geoprocessamento**, que nada mais é do que a exibição de dados em mapas geográficos. Na célula de código abaixo, vamos importar a biblioteca de geração de mapas geográficos. Para isso, vamos fazer uso da biblioteca **Folium**. Para usarmos, precisamos, precisaremos instalá-la através do comando `pip install folium`.

```python
import folium 
```

Na próxima célula de código, vamos gerar um mapa e fazê-lo abrir na cidade de Brasília - DF.

```python
mapa = folium.Map(location = [-15.7834301996, -47.899], zoom_start=13)
display(mapa)
```

Na próxima célula de código vamos criar marcação de lugares.

```python
folium.Marker(location = [-15.7948, -47.91319], popup='SENAI SIG.\nOnde esta acontecendo o curso de Python Developer!',icon=folium.Icon(icon='user',color='black')).add_to(mapa)
folium.Marker(location = [-15.7834301996, -47.899], popup='Estáio Nacional Mané Guarrincha.', icon=folium.Icon(color="red")).add_to(mapa)
display(mapa)
```

Salvando o mapa em uma página HTML:


```python
mapa.save("index.html")
```
